```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D


array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
```

```python
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

plt.figure(figsize=(10,6))
plt.hist(exp_data, bins=100, density=True, histtype='step', color='red', label='Exponential')
plt.hist(uniform_data, bins=100, density=True, histtype='step', color='blue', label='Uniform')
plt.hist(normal_data, bins=100, density=True, histtype='step', color='green', label='Normal')
plt.axis([-2.5, 10, 0, 1.2])
plt.legend()
plt.title('Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('Z = X² + Y²')
plt.show()
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

pearson_corr = df[features].corr(method='pearson')
spearman_corr = df[features].corr(method='spearman')

plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation')

plt.subplot(1,2,2)
sns.heatmap(spearman_corr, annot=True, cmap='YlGnBu')
plt.title('Spearman Correlation')

plt.show()
```
